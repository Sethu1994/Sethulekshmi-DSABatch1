# 5.Perform meaningful visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(Iris_data['SL'],Iris_data['SW'],color='c')
plt.xlabel('SL',fontsize=15)
plt.ylabel('SW',fontsize=15)
plt.show()
```

```python
plt.boxplot(Iris_data['SL'])
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.hist(Iris_data['SW'],color='g',rwidth=0.75)
plt.title('Distribution of SW')
plt.show()
```

```python
plt.scatter(Iris_data['SL'],Iris_data['SW'],s=20,color='r',marker='*')
plt.show()
```
